# Dicionarios

```python

palavras_luis = ['CULTURA', 'SEGURANCA', 'LAZER', 'TRABALHO', 'EDUCACAO', 'ESPORTE', 'SAUDE', 'TRANSPORTE', 'AMBIENTAL', 'SOCIAL']
valores_luis = [16, 16, 17, 17, 18, 18, 18, 21, 23, 24]
d_luis = dict(zip(palavras_luis, valores_luis))

valores_pedro = [4, 4, 4, 4, 4, 4, 5]
palavras_pedro = 'ECONOMIA PUBLICO SOCIAL IMPOSTOS AMBIENTE CRIANCAS MOBILIDADE'.split()
d_pedro = dict(zip(palavras_pedro, valores_pedro))

valores_victor = [1, 2, 3, 5, 6, 6, 6, 13, 16]
palavras_victor = 'AMBIENTE LAZER SEGURANCA TRANSPORTE EDUCACAO EMPRESAS LIXO POPULACAO SAUDE'.split()
d_victor = dict(zip(palavras_victor, valores_victor))

valores_bruno = [15, 17, 22, 22, 27, 29, 35, 44, 62, 64]
palavras_bruno = 'cultura lazer impostos transporte trabalho ambiente seguranca educacao social saude'.split()
d_bruno = dict(zip(palavras_bruno, valores_bruno))
```
