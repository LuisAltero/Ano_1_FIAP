# Questão 3

- – No dataset após o join, obter o índice de Gini por UF (state). Qual UF chamou atenção? Por
que isto ocorreu?

Resposta: O estado do Distrito Federal chamou atenção devido seu valor ser igual a zero, indicando uma igualdade perfeita, segundo o índice de gini. Isso ocorreu devido a existência de apenas um municipio nesse estado(Brasília), fazendo com que a comparação de Brasília fosse feita com ela mesma.Assim, indicando que o Gini não pode ser utilizado com apenas um local para ser analisado.

```python
def gini(array):
  array = array.flatten()
  if np.amin(array) < 0:
    array -= np.amin(array)
  array += 0.0000001
  array = np.sort(array)
  index = np.arange(1,array.shape[0]+1)
  n = array.shape[0]
  return ((np.sum((2 * index - n - 1)* array)) / (n * np.sum(array)))
```

```python
dados_gini_uf = dados_censo_covid.groupby('state')['RDPC'].apply(lambda x: gini(x.values)).to_frame().reset_index()
dados_gini_uf
```

# Questão 4

- Obtenha o sumário das colunas por UF (state), similar a figura abaixo (apresentar o código):

```python
tabela_estado = dados_censo_covid.groupby('state').agg(CasosTotais = pd.NamedAgg('totalCases', 'sum'),
                                                     MortesTotais = pd.NamedAgg('deaths', 'sum'),
                                                     MediaESPVIDA = pd.NamedAgg('ESPVIDA', 'mean'),
                                                     MediaE_AnosESTUDO = pd.NamedAgg('E_ANOSESTUDO','mean'),
                                                     MediaT_ANALF18M = pd.NamedAgg('T_ANALF18M','mean'),
                                                     Mediana_IDHM = pd.NamedAgg('IDHM', 'median'),
                                                     Mediana_RDPC = pd.NamedAgg('RDPC','median'),
                                                     DP_RDPC = pd.NamedAgg('RDPC','std')    
                                                     ).reset_index()
tabela_estado
```
